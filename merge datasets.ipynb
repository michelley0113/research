# Second group of links


```python
# Phantombuster file (pt2)
result_pt2 = pd.read_csv("phantombuster_pt2.csv", encoding = "ISO-8859-1", delimiter = ',', dtype = object)

# replace all commas with periods to put into csv later
result_pt2 = result_pt2.stack().str.replace(',','.').unstack()

# cast links as strings
result_pt2['linkedinProfile'] = result_pt2['linkedinProfile'].astype(str)

# replace NaNs with empty strings
result_pt2 = result_pt2 .replace(np.nan, '', regex = True)
result_pt2  = result_pt2 .replace('nan', '', regex = True)

result_pt2.to_csv('results_cleaned_pt2.csv', quoting = csv.QUOTE_ALL)
```

```python
merged = pd.read_csv('final_with_data.csv', encoding = "ISO-8859-1", delimiter = ',', dtype = object)

# replace NaNs with empty strings
merged = merged.replace(np.nan, '', regex = True)
merged = merged.replace('nan', '', regex = True)
```

```python
# merge result_pt2 and merged on column 'linkedinProfile'

merged = pd.merge(merged, result_pt2, how='left', left_on = 'linkedinProfile', right_on = 'linkedinProfile')
```

```python
columns = [
       'firstName', 'lastName', 'fullName', 'connectionDegree', 'vmid',
       'company', 'companyUrl', 'jobTitle', 'jobDescription', 'location',
       'company2', 'companyUrl2', 'jobTitle2', 'location2', 'school',
       'schoolUrl', 'schoolDateRange', 'school2', 'schoolUrl2',
       'schoolDateRange2', 'baseUrl', 'profileId', 'timestamp', 'description',
       'imgUrl', 'jobDescription2', 'schoolDegree', 'schoolDegreeSpec',
       'schoolDegreeSpec2', 'allSkills', 'skill1', 'endorsement1', 'skill2',
       'endorsement2', 'skill3', 'endorsement3', 'skill4', 'endorsement4',
       'skill5', 'endorsement5', 'skill6', 'endorsement6', 'schoolDescription',
       'schoolDegree2', 'schoolDescription2', 'twitter', 'mail']

for col in columns:
    merged[col] = merged[col + '_x'].map(str) + merged[col + '_y'].map(str)
```

```python
merged = merged.replace('nan', '', regex = True)

for col in columns:
    merged = merged.drop(col + '_x', axis = 1)
    merged = merged.drop(col + '_y', axis = 1)
```

```python
# drop duplicate indexes (only keep first) to correct for inconsistency
merged = merged.drop_duplicates(subset = ['original_index'])

cols_to_drop = ['linkedinSalesNavigatorUrl','timestamp','mutualConnectionsUrl', 'connectionDegree']

# drop irrelevant columns
merged = merged.drop(cols_to_drop, axis = 1)
```

```python
# rename columns for ease of understanding
merged = merged.rename({'fullName':'linkedin_fullName', 'firstName':'linkedin_firstName', 'lastName':'linkedin_lastName'})
```

```python
merged.to_csv('final_results.csv', quoting = csv.QUOTE_ALL)
```
