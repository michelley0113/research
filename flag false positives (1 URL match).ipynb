```python
final_copy = pd.read_csv('final copy.csv', sep = ',')

# replace NaN with empty strings
final_copy = final_copy.replace(np.nan, '', regex = True)

re_search = final_copy[final_copy['linkedinProfile'] == '']
```

```python
# search 'full_name' + 'linkedin' + 'patent'

from googlesearch import search 

re_search['url1'] = ''
re_search['url2'] = ''
re_search['url3'] = ''

def finder(query):
    c = search(query, lang = 'en', num = 3, stop = 3)
    return c

counter = 0
for index, row in re_search.iterrows():
    name = re_search['first_last_name'][index] + ' '
    query = str(name + 'linkedin' + 'patent')
    
    # generator object that finds the top 3 searches
    s = finder(query) 
    
    count = 1
    for i in s:
        # take only top 3 searches
        if count == 4:
            break
            
        re_search['url' + str(count)][index] = i
        count += 1
    
    counter += 1
    
    if counter % 10 == 0:
        print(counter)

# put pause between each search so Google won't block IP
```

```python
count = 0
for index, row in re_search.iterrows():
    # make all urls lowercase
    url1 = str(re_search['url1'][index]).lower()
    url2 = str(re_search['url2'][index]).lower()
    url3 = str(re_search['url3'][index]).lower()
    
    # make last name lower case
    last_name = str(re_search['inventor_name_last'][index]).lower()

    # check first URL, empty if url is not personal profile or last name not in url
    if 'linkedin.com/in/' in url1 and last_name in url1:
        pass
    else:
        re_search['url1'][index] = ""
        
        
    # check second URL, empty if url is not personal profile or last name not in url2
    if 'linkedin.com/in/' in url2 and last_name in url2:
        pass
    else:
        re_search['url2'][index] = ""
        
        
    # check third URL, empty if url is not personal profile or last name not in url3
    if 'linkedin.com/in/' in url3 and last_name in url3:
        pass
    else:
        re_search['url3'][index] = ""
    
    count += 1
    if count % 10 == 0:
        print(count)
```

```python
re_search['test_url'] = re_search['url1']+re_search['url2']+re_search['url3']

re_search = re_search.drop(['Unnamed: 0', 'Unnamed: 0.1','Unnamed: 0.2',
       'linkedinProfile', 'Unnamed: 18', 'Unnamed: 19', 'url1', 'url2', 'url3'], axis = 1)

re_search.to_csv('patent_search.csv', sep = ',')
```
